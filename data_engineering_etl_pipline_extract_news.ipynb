```python
from bs4 import BeautifulSoup
import requests
```

```python
url='https://www.mcqbuddy.com/exam-wise-mcqs/important-mcqs-for-ssc-cgl'
req=requests.get(url)
content=req.text
```

```python
soup=BeautifulSoup(content)
data=soup.find_all('a',{'class':'page-link lightlink'})
data=data[len(data)-1]
max_page_size=data['href'].split('page=')[1]
QUESTIONS=[]
META_DATA=[]
```

```python
for i in range(1,int(max_page_size)):
    URL=url+'?page={}'.format(i)
    req=requests.get(URL)
    content=req.text
    soup=BeautifulSoup(content)
    print (URL)
    
    data=soup.find_all('div',{'class':'questions'})
    for ques in data:
        QUESTIONS.append(ques.text.split('\n')[1])
        
        
    data=soup.find_all('div',{'class':'options'})
    for ans in data:
        option=ans.find_all('div')
        a=[]
        for i in option:
            main_ans=i['data-correct']
            a.append(i.text)
        META_DATA.append({
            'ans':main_ans,
            'options':a
        })
```

```python
DATA=[]
for question,meta in zip(QUESTIONS,META_DATA):
    temp=[]
    temp.append(question)
    for i in meta['options']:
        temp.append(i)
    temp.append(meta['ans'])
    DATA.append(temp)
```

```python
import pandas as pd
```

```python
data=pd.DataFrame(DATA,columns=('Question','Option','Option','Option','Option','Answer'))
```
